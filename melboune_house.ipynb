```python
# Load the first CSV file
df = pd.read_csv('/content/MELBOURNE_HOUSE_PRICES_LESS.csv')

# Load the second CSV file
df_train = pd.read_csv('/content/Melbourne_housing_FULL.csv')

# Merge the two dataframes using the concat function
merged_df = pd.concat([df, df_train])

# Write the merged dataframe to a new CSV file
merged_df.to_csv('merged_file.csv', index=False)
```

```python

# Skewness is a measure of symmetry, or more precisely, the lack of symmetry.
# A distribution, or data set, is symmetric if it looks the same to the left and right of the center point. 
# Kurtosis is a measure of whether the data are heavy-tailed or light-tailed relative to a normal distribution.

import warnings
warnings.filterwarnings("ignore")

#kernel density plot
sns.distplot(merged_df.Price)
plt.show()
print('\n')
print("skewness: %f" % merged_df['Price'].skew())
print("kurtosis: %f" % merged_df ['Price'].kurt())
     
```

```python
# log transformations used to transform skewed data to approximately conform to normality.
# to know more see note attached

# log transform the target 
merged_df["Price"] = np.log1p(merged_df["Price"])

#Kernel Density plot
sns.distplot(merged_df.Price);
plt.show()
print('\n')
print("skewness: %f" % merged_df['Price'].skew())
print("kurtosis: %f" % merged_df ['Price'].kurt())
```

```python
num=merged_df[['Rooms','Price','Postcode','Propertycount','Distance']]
num.head()
```

```python
num.fillna(method='ffill', inplace=True)
     
```

```python
cat=merged_df[['Type','Method','Regionname','CouncilArea']]
cat.head()
```

```python
cat.fillna(merged_df.mode().iloc[0], inplace=True)
     
```

```python
cat_data=pd.DataFrame((cat),columns=cat.columns)
cat_data.head()
```

```python
cat_data.isnull().sum()
```

```python
cat_data.nunique()
```

```python
df_dummies =pd.get_dummies(cat_data)
df_dummies
```

```python
final_data=num.join(df_dummies)
final_data.head()
```

```python
final_data.shape
```

```python
# Reset the index
final_data = final_data.reset_index(drop=True)
```

```python
# Split the data into features and target variable
X = final_data.drop('Price', axis=1)
y = final_data['Price']
     

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train,y_train)
```

```python
trainpreds_linreg=linreg.predict(X_train)
```

```python
linreg.fit(X_test,y_test)
testpreds_linreg=linreg.predict(X_test)

# Error Metrics
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mae = mean_absolute_error(y_test, testpreds_linreg)
mse = mean_squared_error(y_test, testpreds_linreg)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, testpreds_linreg)
print("Mean Absolute Error:",mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```
